```python
import seaborn as sns
```

_tips dataset - already existing in seaborn_

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='total_bill',y="tip",data=tips) #bivariate
```

### relationship
- univariate : single variable
- bivariate : two variables
- multivariate : multiple variables


```python
sns.scatterplot(x='size',y="tip",data=tips) #bivariate
```

```python
sns.lineplot(x='size',y="tip",data=tips) 
```

```python
sns.distplot(tips['tip']) #univariate
```

### barplot 
> numerical data vs categorical data

```python
sns.barplot(x='time',y='tip',data=tips)
```

```python
sns.boxplot(x='size',y='tip',data=tips)
```

```python
tips.describe()
```

```python
sns.catplot(x='smoker',y='tip',data=tips) #categorical vs numerical
```

```python
sns.catplot(x='day',y='tip',data=tips)
```

```python
sns.countplot(x='day',data=tips)
```

```python
tips.corr
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True)
```

> correlation is between -1 to 1
 if it is greater than 0.5 , they are highly correlated
 if it is between -0.5 to 0.5 then no significance
 if it is less than -0.5 then highly inverse proportion

```python
sns.jointplot(x='tip',y='total_bill',data=tips) # histogram and scatter plot
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips, kind = 'hex')
```
