```python
#replace null with either mode, median or mean
#use mode when we have categories, here geography and gender
#for small range numerical data use mean, here age
#for large range numerical data use median, here balance
```

```python
df['Geography'].fillna(df['Geography'].mode()[0],inplace=True) #inplace=true makes changes permanent
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True) #inplace=true makes changes permanent
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True) #inplace=true makes changes permanent
```

```python
df['Balance'].fillna(df['Balance'].median(),inplace=True) #inplace=true makes changes permanent
```

```python
#convert textual data to numerical using label encoding
#label encoding works in alphabetical order so France will be 0,germany will be 1, spain will be 2
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Geography'] = le.fit_transform(df['Geography'])
df['Gender'] = le.fit_transform(df['Gender'])
```
