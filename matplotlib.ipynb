```python
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([-7,-6,-5,-4,-3,-2,-1,1,2,3,4,5,6,7])
y = np.power(x,3)
plt.plot(x,y) #simple line plot
```

```python
plt.scatter(x,y,color='yellow')
```

```python
plt.scatter(x,y,color='r',marker=">") # > or < or s or * or .
```

```python
plt.scatter(x,y,color='r',marker=".") 
plt.title("relation between x and y")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
plt.plot(x,y) #simple line plot
plt.title("different")
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y,'go')
plt.title("y=sin x")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25,5.25],[50,40,60,20,100,12],color="red",label="students",width=0.5)
plt.xlabel("students")
plt.ylabel("marks")
plt.legend()
plt.title("marks of students")
```

```python
name = ["messi","ronaldo","mbappe","haaland","neymar"]
age=[35,37,24,22,33]
plt.bar(name,age)
```

```python
plt.barh(name,age)
plt.xlabel("age")
```

```python
plt.figure(figsize=(10,8)) #width and height
x = np.linspace(0,10,1000)
y = np.sin(x)
y1 = np.cos(x)
plt.subplot(2,3,1)
plt.plot(x,y,'y')
plt.subplot(2,3,6)
plt.plot(x,y1,'r')
```

```python
#stacked bar
x=name
y=age
y1=[22,35,24,35,46]
plt.bar(x,y1,label="a",width=0.5,color='r')
plt.bar(x,y,label="b",width=0.5,color='g')
```

```python
plt.barh(x,y1,label="a",color='r')
plt.barh(x,y,label="b",color='g')
```

### piechart

```python
marks=[48,35,27,46]
subjects=["maths","physics","english","python"]
colors=['g','r','y','b','o']
plt.pie(marks,labels=subjects,colors=colors)
```

```python
marks=[48,35,27,46]
subjects=["maths","physics","english","python"]
colors=['g','r','y','b','o']
plt.pie(marks,labels=subjects,colors=colors,autopct="%1.1f%%")
```

```python
marks=[48,35,27,46]
subjects=["maths","physics","english","python"]
colors=['g','r','y','b','o']
plt.pie(marks,labels=subjects,colors=colors,autopct="%1.1f%%",startangle=90)
```

```python
marks=[48,35,27,46]
subjects=["maths","physics","english","python"]
colors=['g','r','y','b','o']
plt.pie(marks,labels=subjects,colors=colors,autopct="%1.1f%%",startangle=90,explode=(0,0,1,0.3))
```

### area plot

```python
x= np.arange(30)
y=np.random.normal(10,11,30) # 10 is mu ans 11 is sigma
y=np.square(y)
plt.plot(x,y,'g')
plt.fill(x,y)
```
